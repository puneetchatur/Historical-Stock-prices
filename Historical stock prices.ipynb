```python
def get_stock_data(symbol, start_date, end_date):
    
    #Creating a ticker object
    tickerData = yf.Ticker(symbol)

    #get the historical prices for this ticker (period='1d' implies daily data)
    tickerDf = tickerData.history(start_date='1d', start=start_date, end=end_date)
    
    if not tickerDf.empty:
        tickerDf.index = tickerDf.index.date
    tickerDf.index.names = ['Date']
    return tickerDf
```

Use the function named

that would take the stock symbol one at a time from the nifty50.csv file to generate a data frame of the above kind and export it as a CSV file inside a NEW FOLDER in the given directory with the name of the file as exactly equal to the symbol of the stock given in the CSV file.

Note:

The symbols given in the dataset do not have the suffix .NS which needs to be added to the symbols before it can be used in the function.

The new folder must be created with the name NIFTY50

```python
import pandas as pd
nifty50 = pd.read_csv('nifty50.csv')
nifty50.head()
```

```python
nifty50.shape
```

```python
def export_stock_data(symbol, start_date, end_date,directory):
    
    #Creates directory if not exists
    if not os.path.exists(directory):
        os.mkdir(directory)
    
    #Creating a ticker object
    tickerData = yf.Ticker(symbol+'.NS')

    #get the historical prices for this ticker (perid='1d' implies daily data)
    tickerDf = tickerData.history(start_date='1d', start=start_date, end=end_date)
    
    #If no historical data is found for the given time range, index processing is skipped
    if not tickerDf.empty:
        tickerDf.index = tickerDf.index.date
        
    tickerDf.to_csv(directory + '/' + symbol + '.csv')
```

```python
#List of symbols
symbols = nifty50['Symbol'].to_list()
```

```python
#Iterate through each symbol and get the data for the given time range
for sym in symbols:
    export_stock_data(symbol=sym, start_date='2019-10-19', end_date='2022-10-19', directory='NIFTY50')
```

**Question 4**

Create a function called,

daily_return(df)

This function should be able to take a dataframe shown in the example above and calculate the daily return of the close price and output a dataframe with the same number of columns and an additional column named "Return_Close".

The daily return for a day is calculated as follows:

daily return = (today's price - previous day's price)/(previous day's price)*100

Note: You can calculate the daily return starting from the day 2. The return for the day 1 should be set to 0.

```python
def daily_return(df):
    #subtract-the-previous-row-from-the-current-row-in-a-pandas-dataframe-and assign the value to a new column
    df['Return_Close'] = df['Close'].pct_change() * 100
    #First row is Nan, so we're replacing with 0
    df.fillna(0,inplace=True)
    return df
```

**Question 5**

Consider the NIFTY50 folder created in Q3. This folder should contain 50 data frames - one for each stocks in the Nifty-50 list. Create a function called,

get_return(directory)


This function would take the link to the NIFTY50 folder, read one CSV file at a time present inside it, calulate the return of the close price, save the return value in an new column called "Return" and update the CSV file.

```python
def get_return(directory):
    
    #List all files in the directory
    file_list = os.listdir(directory)
    
    #Iterate through each file
    for file in file_list:
        # reading the file 
        df = pd.read_csv(directory + '/' + file)
        
        #This check is given if there are no records for the given time range for a given company
        if not df.empty:
            #subtract-the-previous-row-from-the-current-row-in-a-pandas-dataframe-and-assign-to-a-new-column
            df['Return'] = df['Close'].pct_change() * 100
            #First row is Nan, so replacing with 0
            df.fillna(0,inplace=True)
            
            #Update csv
            df.to_csv(directory + '/' + file)
```

```python
get_return(directory='NIFTY50')
```

**Question 6**

Modify the get_return() function such that it has another argument,

get_return(directory, plot = False)

If the argument plot is set to False the the output of this function is same as the function in Q6. However, if it is set to True then for every stock we get a line plot in the JPEG format exported in a separate folder named Graphs inside the parent directory

```python
def get_return(directory,plot):
    
        
    #Creates directory if not exists
    if plot==True:
        if not os.path.exists('Graphs'):
            os.mkdir('Graphs')

    #List all files in the directory
    file_list = os.listdir(directory)

    #Iterate through each file
    for file in file_list:
        #print(file)
        #Read the file
        df = pd.read_csv(directory + '/' + file)

        df.rename(columns={'Unnamed: 0':'Date'},inplace=True)

        #This check is given if there are no records for the given time range for a given company
        if not df.empty:
            #https://stackoverflow.com/questions/39479919/how-do-i-subtract-the-previous-row-from-the-current-row-in-a-pandas-dataframe-an
            df['Return'] = df['Close'].pct_change() * 100
            #First row is Nan, so replacing with 0
            df.fillna(0,inplace=True)
            
            #Update csv
            df.to_csv(directory + '/' + file)
            
            if plot:

                df.plot.line(x = 'Date', y = 'Return')
                plt.xlabel('Date')
                plt.ylabel('Return')
                # Rotates X-Axis Ticks by 45-degrees
                plt.xticks(rotation = 25)
                #Remove .csv from the file name
                plt.title(file.split('.')[0])
                #plt.show()

                plt.savefig('Graphs'+ '/' + file.split('.')[0] + '.jpg')
```
